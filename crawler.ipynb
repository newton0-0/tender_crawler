```python
import requests
from bs4 import BeautifulSoup

# Function to extract a specific URL based on link text
def get_specific_link(url, target_text):
    # Send a GET request to the website
    response = requests.get(url)
    
    # Check if the request was successful
    if response.status_code == 200:
        # Parse the page content
        soup = BeautifulSoup(response.content, 'html.parser')
        
        # Find all links (anchor tags)
        links = soup.find_all('a', href=True)
        
        # Loop through the links to find the one with the target text
        for link in links:
            if link.get_text(strip=True) == target_text:
                return link.get('href')
        return f'Link text "{target_text}" not found.'
    else:
        return f"Failed to retrieve the page. Status code: {response.status_code}"
    
# Function to extract data from a webpage
def web_scrape(url):
    # Send a GET request to the website
    response = requests.get(url)
    
    # Check if the request was successful
    if response.status_code == 200:
        # Parse the page content
        soup = BeautifulSoup(response.content, 'html.parser')
        
        # Example: Find and print all the titles on the page
        titles = soup.find_all('h1')  # Adjust the tag as per your need
        
        for index, title in enumerate(titles, start=1):
            print(f'Title {index}: {title.get_text(strip=True)}')
        
        # Extract more data by finding elements using CSS selectors or tags
        # Example: Extracting all links
        links = soup.find_all('a', href=True)
        for link in links:
            print(f'Link text: {link.get_text(strip=True)}, URL: {link["href"]}')
    else:
        print(f"Failed to retrieve the page. Status code: {response.status_code}")

# URL of the website you want to scrape
url = "https://etenders.gov.in/eprocure/app"

# The specific link text you want to extract
target_text = "Tenders by Organisation"

# Extract and print the specific link
org_url = get_specific_link(url, target_text)
result = web_scrape("https://etenders.gov.in" + org_url + "?")
print(result)
```
